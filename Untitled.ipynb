```python
import pandas as pd
import matplotlib.pyplot as plt
from cassandra.cluster import Cluster
from IPython.display import display


# Connect to Cassandra (use localhost if inside the same container)
cluster = Cluster(['172.19.0.3'])
session = cluster.connect('youtube')

# Query trending videos
rows = session.execute("""
    SELECT video_id, title, views, publish_time
    FROM trending_videos
""")

# Convert to DataFrame
data = [(row.video_id, row.title, row.views, row.publish_time) for row in rows]
df = pd.DataFrame(data, columns=["video_id", "title", "views", "publish_time"])

# Convert publish_time to datetime if not already
df["publish_time"] = pd.to_datetime(df["publish_time"])

# Sort DataFrame by publish time
df = df.sort_values("publish_time")

# Display title
print("📊 YouTube Trending Analytics")

# Plot total views over time
plt.figure(figsize=(10, 5))
df.groupby("publish_time")["views"].sum().plot(marker='o')
plt.xlabel("Publish Time")
plt.ylabel("Total Views")
plt.title("Total Views Over Time")
plt.grid(True)
plt.tight_layout()
plt.show()

# Show raw data (you can choose to show manually)
display(df.head(10))  # You can change to `df` to show all
```
